```python
new_set = df[df['CLASS'].isin(['A', 'C', 'E', 'F', 'G'])]
new_set.shape
#(101456, 38) After filtering out only the scheduled passenger class
#We have 101,456 records to work with. NOTE: this is only 2023 data
new_set.describe()
new_set.head(10)
new_set['CARRIER_NAME']
```

```python
#Since this dataset doesn't have a marketshare column, which to be used as a label 
#We have to create one: 
#Compute total passengers for each carrier: 
total_flights = new_set.groupby('CARRIER_NAME')['UNIQUE_CARRIER_NAME'].size()
#compute market share for each carrier 
market_share = total_flights / total_flights.sum()
#add as a column 
df = new_set.merge(market_share.rename('CARRIER_SHARE'), left_on='CARRIER_NAME', right_index=True)
df['LOG_CARRIER_SHARE'] = np.log(df['CARRIER_SHARE'] + 1)
```

```python
# Splitting Data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_x, df_y, test_size = 0.3, random_state=42)
```

```python
#Run linear regression
from sklearn.linear_model import LinearRegression

# S2.1, define the linear regression function (plug in x^0, x, x^2, x^3, then obtain betas)

lr = LinearRegression()

## S2.2, train the model

lr.fit(X_train,y_train)

y_test_pred = lr.predict(X_test)

# Get error, square it, get the mean
e = y_test - y_test_pred

MSE_test = np.mean(e**2)

print(MSE_test)
```

```python
lr.score(X_test,y_test)
```

```python
#Run the normal poly regression model: 
from sklearn.preprocessing import PolynomialFeatures

poly3 = PolynomialFeatures(degree = 3)
```

```python
X_poly_train = poly3.fit_transform(X_train)

print(X_poly_train.shape)
```

```python
#Run linear regression
from sklearn.linear_model import LinearRegression

# S2.1, define the linear regression function (plug in x^0, x, x^2, x^3, then obtain betas)

lr = LinearRegression()

## S2.2, train the model

lr.fit( X_poly_train  , y_train)
```

```python
# S3. Predict and calculate error

## Calculate mse
# lr.predict(X_test) => transfer X_test also
X_test.shape
X_poly_test = poly3.fit_transform(X_test)

y_test_pred = lr.predict(X_poly_test)

# Get error, square it, get the mean
e = y_test - y_test_pred

MSE_test = np.mean(e**2)

print(MSE_test)
#considering the values for log market share is low -> this is a low MSE
```

```python
# Calculate the variance of the actual values
variance_y = np.var(y_test)

# Calculate R-squared
R_squared = 1 - (MSE_test / variance_y)

print(f"R-squared: {R_squared}")
```

```python
### In below, use make_pipeline for practice
from sklearn.preprocessing  import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# S1. Define models - two models, polynomial transfer first, then linear regression 

pipe_poly3 = PolynomialFeatures(degree = 3)
pipe_lr = LinearRegression()

poly3_reg = make_pipeline(pipe_poly3,pipe_lr)

# S2. Apply polynomial regression in pipeline

poly3_reg.fit(X_train, y_train)

# S3. Predict and get mse
y_test_pred_pipe = poly3_reg.predict(X_test) #polynomical transfer will be taken care inside the pipeline

np.mean((y_test - y_test_pred_pipe)**2)
```

```python
#Run grid
from sklearn.model_selection import GridSearchCV

# Load other modules
from sklearn.preprocessing  import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression

# S1. Define grid of parameter. 
## The name should be exactly the same, otherwise cannot find which to specify.
param_poly = {'polynomialfeatures__degree' :  range(1,6)  } # range(i,j): consec.integer i-j, exclude j
## Dictionary format to define. 
my_param_poly = {'polynomialfeatures__degree'  : [1, 2, 3, 4, 5] }
## Two methods are equivalent

# S2. Define estimator: use make_pipeline to combine two functions. 
pipe_poly = PolynomialFeatures()
pipe_lr = LinearRegression()
poly_reg = make_pipeline( pipe_poly , pipe_lr)
# S3. Define GridSearchCV Estimation function, then train the ,model
grid_poly = GridSearchCV(poly_reg, param_poly, cv = 5)
grid_poly.fit(X_train, y_train)

```

```python
print(grid_poly.score(X_test, y_test)) #score = 0.8793
print(grid_poly.score(X_train, y_train)) #0.8772
```

```python
y_pred_grid_test = grid_poly.predict(X_test)
y_test - y_pred_grid_test
# Get error, square it, get the mean
e = y_test - y_pred_grid_test

MSE_test = np.mean(e**2)
print(MSE_test) #0.000642
#with the scores and the MSE score, this seems like a good model. I will stop here
```

```python
from sklearn.model_selection import GridSearchCV

# 1. Define a list of parameters (key is 'alpha')

lasso_params = { 'alpha' :  [0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 10] }
 
# 2. Define function and fit the data

lasso = Lasso()

lasso_grid = GridSearchCV(lasso, lasso_params, cv = 5)

lasso_grid.fit(X_train, y_train)

# 3.1 Present performance measure

## 3.1a: unbiased performance measure in R2

print(lasso_grid.score(X_test, y_test))
```

```python
## 3.1b: performance measure based on which hyperparameter is chosen (R2)

print(lasso_grid.best_score_)

# 3.2 find best hyperparameters

print(lasso_grid.best_params_)

# 3.3 find best parameter estimates

# First get optimal model

lasso_grid.best_estimator_.coef_, lasso_grid.best_estimator_.intercept_

```

```python
from sklearn.metrics import r2_score, mean_squared_error

lasso_best = lasso_grid.best_estimator_
y_pred = lasso_best.predict(X_test)

r2 = r2_score(y_test, y_pred)
print(f"R-squared: {r2}")

mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```
